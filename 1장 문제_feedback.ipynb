# 1-3 데이터 변환

## Q1

다음 식이 나타내는 변환에 대한 설명으로 가장 옳지 않은 것은 무엇인가요? 

$X_{new} = \begin{cases} 
      X^\lambda & \text{if } \lambda \neq 0 \\
      \log(X) & \text{otherwise}
   \end{cases}$

1) X가 음수일 때는 적용할 수 없습니다.

2) $\lambda$ 가 1보다 크면 $X_{new}$의 분포는 $X$ 보다 좌측으로 치우치게 됩니다.

3) $\lambda$ 가 1보다 작으면 $X_{new}$는 $X$의 분포보다 우측의 꼬리 영역의 길이가 짧게 됩니다.

4) $\lambda$ 가 음수라면 $X_{new}$ 값은 $X$ 보다 작아 집니다.

<font color='red'> 정답 4번: $\lambda$가 음수 일경우 $X$ 값에 따라 작아질 수도 있고 커질 수도 있습니다. </font>

# Q2


다음 변수 처리에 대한 설명으로 옳지 않은 것은 무엇인가요?

1) 정규화는 데이터의 범위를 일치 시키는 과정으로, 주로 0과 1사이의 값으로 스케일을 조정합니다.

2) 표준화는 데이터를 평균이 0이고 표준 편차가 1인 분포로 변환하는 과정입니다.

3) 가변수는 범주형 변수를 이진형 변수로 변환하는 과정으로 범주수 만큼의 가변수를 생성하여, 범주의 해당 여부를 1과 0으로 나타내게 합니다.

4) 박스콕스 변환을 거친 변수는 항상 정규 분포를 따르게 됩니다.


<font color='red'> 정답 4번: 박스콕스 변환은 정규분포에 가까운 형태로 변환시켜 줄 뿐 정규 분포에 따르는 것은 보장하지 않습니다.</font>
